```python
import requests
import pandas as pd
```

```python


headers = {
    'sec-ch-ua': '^\\^Not',
    'sec-ch-ua-platform': '^\\^Windows^\\^',
    'Referer': 'https://www.ebooks.com/en-bd/subjects/architecture/',
    'sec-ch-ua-mobile': '?0',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36',
    'Content-Type': 'application/json',
}

params = (
    ('CountryCode', 'BD'),
    ('subjectId', '470'),
    ('pageNumber', '1'),
)

response = requests.get('https://www.ebooks.com/api/search/subject/', headers=headers, params=params)
```

```python
title = []
edition = []
subtitle = []
authors = []
publisher = []
publication_year = []
price = []
description = [] 
short_publication_date = []

for results in result_json:
    title.append(results['title'])
    edition.append(results['edition'])
    subtitle.append(results['subtitle'])
    authors.append(results['authors'])
    publisher.append(results['publisher'])
    publication_year.append(results['publication_year'])
    price.append(results['price'])
    description.append(results['description'])
    short_publication_date.append(results['short_publication_date'])
```

```python
books_df = pd.DataFrame({'Title':title,'Edition':edition,'Subtitle':subtitle,'Author':authors,'Publisher':publisher,'Publication Year':publication_year,'Price':price,'Description':description,'Publication Date':short_publication_date})
```
